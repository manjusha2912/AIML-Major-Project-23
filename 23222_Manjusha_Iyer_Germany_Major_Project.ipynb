To identify the extreme Price values,

calculate the z-scores for the price column by subtracting the mean and dividing by the standard deviation.
Set a threshold (in this case, >3 and <-3) to identify outliers.
The records with a z-score greater than 3 or less than-3 are considered as price outliers

```python
# calculate the z-scores - rerun step 3

usedcars['Price_zscore'] = (usedcars['price'] - usedcars['price'].median()) / usedcars['price'].std()

outlier_recs = usedcars[usedcars['Price_zscore']>3]
sorted_outlier_recs = outlier_recs.sort_values(by='Price_zscore', ascending=False)

print(sorted_outlier_recs)
```

The range of Price_zscore (3.01,61.11) indicates existence of data-skewness and right-tail graph is displayed in the price distribution chart. Among these 704 rows of price-outliers, the data row with a price value of 119990 is an outright extreme value.



```python
#rerun step 4
usedcars_clean=usedcars[usedcars['Price_zscore'] <= 3]
```

```python
# Group the data by car model
grouped = usedcars_clean.groupby(['make'])
#usedcars.groupby(['fuel'])['fuel'].count().sort_values(ascending=False)

# Define a function to identify outliers within a group
def find_outliers(group):
    q1 = group['hp'].quantile(0.25)
    q3 = group['hp'].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = group[(group['hp'] < lower_bound) | (group['hp'] > upper_bound)]
    print(outliers)
    return outliers

# Create a DataFrame to store the outliers for each car model
outliers_per_model = usedcars_clean[['make','model','hp']]
print(outliers_per_model)

# Iterate over each model group, find outliers, and append to the result DataFrame
for model, group in grouped:
    outliers = find_outliers(group)
    outliers_per_model = pd.concat([outliers_per_model, outliers])

# Display the outliers per car model
#print(outliers_per_model)
```

Classify the car dataset on consumer defined luxuary segments

```python
# rerun step 8
def classify_hp_to_luxury(hp):
    if hp <= 150:
        return "L01:Entry-Level"
    elif hp <= 250:
        return "L02:Mid-Size"
    elif hp <= 350:
        return "L03:Full-Size"
    elif hp <= 400:
        return "L04:Sports-Performance"
    elif hp <= 450:
        return "L05:Sports-UVs"
    elif hp <= 500:
        return "L06:Coupes-Convertibles"
    elif hp <= 550:
        return "L07:Ultra-Luxury"
    elif hp <= 600:
        return "L08:Electric-Hybrid"
    elif hp <= 650:
        return "L09:Sports-Luxury"
    elif hp <= 700:
        return "L10:Grand-Tourers"
    else:
        return "L11:Super-Premium"

usedcars_clean['luxury_segment'] = usedcars_clean['hp'].apply(classify_hp_to_luxury)
print(usedcars_clean.head(10))
```

LINEAR REGRESSION MODEL

```python
#Linear Regression Model

lr = LinearRegression()
lr.fit(X_train,y_train)

# Get coefficients and intercept
coefficients = lr.coef_
intercept = lr.intercept_

# Display equation of the model
equation = f"y = {intercept:.2f} + "
for i, coeff in enumerate(coefficients):
    equation += f"{coeff:.2f} * {X_test.columns[i]} + "
equation = equation[:-2]  # Remove the last '+'
print("Equation:", equation)

print("Linear Regression Model")
print(f"Train Score: {lr.score(X_train, y_train):.2f}")
print(f"Test Score: {lr.score(X_test, y_test):.2f}")
```

Assess the proportions of upper and lower bias in the predicted values

```python
# bias distribution

y_predict = xgb_model.predict(X_test)

upper_bias_count = 0
lower_bias_count = 0

for true_val, pred_val in zip(y_test, y_predict):
    if pred_val > true_val:
        upper_bias_count += 1
    elif pred_val < true_val:
        lower_bias_count += 1

total_samples = len(y_test)
upper_bias_proportion = upper_bias_count / total_samples
lower_bias_proportion = lower_bias_count / total_samples

print("Total Predicted records: ",  total_samples)
print(f"Proportion of Upper Bias:  {upper_bias_proportion:.2f}")
print(f"Proportion of Lower Bias:  {lower_bias_proportion:.2f}")
```
